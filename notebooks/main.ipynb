```python
# Loading library
import numpy as np
import pandas as pd
```

```python
fake = pd.read_csv("data/Fake.csv")
fake.head()
fake["label"] =0
fake.head()
```

```python
true = pd.read_csv("data/True.csv")
true.head()
true["label"] =1
true.head()
```

```python
news = pd.concat([fake, true], axis=0, ignore_index=False)
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum() 
```

```python
news.drop(["title", "subject","date"], axis=1, inplace=True)
news.head()
```

```python
news = news.sample(frac=1) #Reshuffle the dataset
news.head()
```

```python
news.reset_index(inplace=True)
news.head()
```

```python
news.drop(["index"], axis=1, inplace=True)
news.head()
```

```python
import re
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r"https?://\S+www\.\S+", " ", text)  # Remove URLs
    text = re.sub(r"<.*?>", " ", text)  # Remove HTML tags
    text = re.sub(r"[^\w\s]", " ", text)  # Remove punctuation
    text = re.sub(r"\d", " ", text)  # Remove digits
    text = re.sub(r"\n", " ", text)  # Remove newlines
    
    return text
```

```python
news["text"] = news["text"].apply(wordopt)
news["text"].head()
```

```python
x = news["text"]
y = news["label"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
xv_train = vectorizer.fit_transform(x_train)
```

```python
xv_test = vectorizer.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(xv_train, y_train)
```

```python
predict_lr = lr.predict(xv_test)
```

```python
lr.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predict_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
dct = DecisionTreeClassifier()
```

```python
dct.fit(xv_train, y_train)
```

```python
predict_dct = dct.predict(xv_test)
```

```python
dct.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_dct))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(xv_train, y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(xv_train, y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
```

```python
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    else:
        return "True News"
```

```python
def manualPredict(news):
    testing_news = {"text": [news]}
    news_def_test = pd.DataFrame(testing_news)  
    news_def_test["text"] = news_def_test["text"].apply(wordopt)
    news_x_test = news_def_test["text"]
    news_xv_test = vectorizer.transform(news_x_test)
    predict_lr = lr.predict(news_xv_test)
    # predict_dct = dct.predict(news_xv_test)
    pred_gbc = gbc.predict(news_xv_test)
    predict_rfc = rfc.predict(news_xv_test)
    return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
    output_label(predict_lr[0]), 
    output_label(pred_gbc[0]), 
    output_label(predict_rfc[0])
)
```

```python
news_article = str(input("Enter the news article: "))
```

```python
manual_testing = manualPredict(news_article)
print(manual_testing)
```
